```python
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('heart_disease_health_indicators.csv')
```

### Univariate Analysis

```python
# a. Examine Distributions of Individual Variables
```

```python
# Plot histogram for Age
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], kde=True, bins=30, color='skyblue')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot histogram for BMI
plt.figure(figsize=(10, 6))
sns.histplot(df['BMI'], kde=True, bins=30, color='lightgreen')
plt.title('Distribution of BMI')
plt.xlabel('BMI')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot histogram for Mental Health Days (MentHlth)
plt.figure(figsize=(10, 6))
sns.histplot(df['MentHlth'], kde=True, bins=30, color='salmon')
plt.title('Distribution of Mental Health Days')
plt.xlabel('Mental Health Days')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot histogram for Physical Health Days (PhysHlth)
plt.figure(figsize=(10, 6))
sns.histplot(df['PhysHlth'], kde=True, bins=30, color='purple')
plt.title('Distribution of Physical Health Days')
plt.xlabel('Physical Health Days')
plt.ylabel('Frequency')
plt.show()
```

```python
# b. Investigate Prevalence of Health Conditions
```

```python
# Define categorical columns
categorical_cols = ['HeartDiseaseorAttack', 'HighBP', 'HighChol','Diabetes']

# Plot count plots for categorical variables
for col in categorical_cols:
    plt.figure(figsize=(8, 6))
    sns.countplot(x=col, data=df, palette='pastel')
    plt.title(f'Distribution of {col}')
    plt.xlabel(f'{col}')
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.show()
```

```python
# c. Analyze Distribution of Heart Disease (Target Variable)
```

```python
# Plotting count plot for 'HeartDisease' column
plt.figure(figsize=(10, 6))
sns.countplot(x='HeartDiseaseorAttack', data=df, palette='coolwarm')
plt.title('Distribution of HeartDiseaseorAttack Cases')
plt.xlabel('HeartDiseaseorAttack')
plt.ylabel('Count')
plt.show()
```

```python
# Plot pie chart for Heart Disease (HeartDisease)
plt.figure(figsize=(8, 8))
df['HeartDiseaseorAttack'].value_counts().plot.pie(autopct='%1.1f%%', colors=['skyblue', 'lightcoral'], labels=['No', 'Yes'])
plt.title('Distribution of HeartDiseaseorAttack Cases')
plt.ylabel('')
plt.show()
```

### Bivariate Analysis

```python
# a. Explore Relationships with Heart Disease

# Compare distribution of key variables between individuals with and without heart disease
```

```python
# Plotting for High Blood Pressure (HighBP)
plt.figure(figsize=(10, 6))
sns.countplot(x='HighBP', hue='HeartDiseaseorAttack', data=df, palette='pastel')
plt.title('Distribution of High Blood Pressure by HeartDiseaseorAttack')
plt.xlabel('High Blood Pressure')
plt.ylabel('Count')
plt.legend(title='HeartDiseaseorAttack', labels=['No', 'Yes'])
plt.show()
```

```python
# Plotting for High Cholesterol (HighChol)
plt.figure(figsize=(10, 6))
sns.countplot(x='HighChol', hue='HeartDiseaseorAttack', data=df, palette='muted')
plt.title('Distribution of High Cholesterol by HeartDiseaseorAttack')
plt.xlabel('High Cholesterol')
plt.ylabel('Count')
plt.legend(title='HeartDiseaseorAttack', labels=['No', 'Yes'])
plt.show()
```

```python
# Plotting for BMI (Body Mass Index)
plt.figure(figsize=(10, 6))
sns.boxplot(x='HeartDiseaseorAttack', y='BMI', data=df, palette='Set3')
plt.title('BMI Distribution by HeartDiseaseorAttack')
plt.xlabel('HeartDiseaseorAttack')
plt.ylabel('BMI')
plt.show()
```

```python
# b. Visualize Correlations Between Variables
```

```python
# Correlation heatmap for continuous variables
plt.figure(figsize=(10, 8))
sns.heatmap(df[['BMI', 'MentHlth', 'PhysHlth', 'Age']].corr(), annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Heatmap of Continuous Variables')
plt.show()
```

```python
# c. Compare Heart Disease Across Demographic Groups

# Analyze prevalence of heart disease across different demographic groups
```

```python
# Pivot table to compare heart disease by gender and education level (example)
heart_disease_pivot = pd.pivot_table(df, index='Sex', columns='Education', values='HeartDiseaseorAttack', aggfunc='mean')
plt.figure(figsize=(10, 6))
sns.heatmap(heart_disease_pivot, annot=True, cmap='YlGnBu', vmin=0, vmax=1)
plt.title('HeartDiseaseorAttack Prevalence by Gender and Education')
plt.xlabel('Education Level')
plt.ylabel('Sex')
plt.show()
```

### Bar Plots and Pie Charts

```python
#  a. Visualize Categorical Variables:

# Define categorical variables
categorical_vars = ['Smoker', 'PhysActivity', 'Fruits', 'Veggies']

# Plotting
plt.figure(figsize=(14, 10))
for i, var in enumerate(categorical_vars, 1):
    plt.subplot(2, 2, i)
    df[var].value_counts().plot(kind='bar', color='teal')
    plt.title(f'Distribution of {var}')
    plt.xlabel(var)
    plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
# b. Compare Heart Disease Across Categories:
```

```python
# Grouped bar charts for comparing heart disease across categories

# Define categorical variables of interest
categories_of_interest = ['Smoker', 'PhysActivity', 'Fruits', 'Veggies']

# Create grouped bar charts
plt.figure(figsize=(14, 10))
for i, var in enumerate(categories_of_interest, 1):
    plt.subplot(2, 2, i)
    grouped_data = df.groupby(var)['HeartDiseaseorAttack'].value_counts(normalize=True).unstack()
    grouped_data.plot(kind='bar', stacked=True, ax=plt.gca(), color=['skyblue', 'salmon'])
    plt.title(f'Heart Disease by {var}')
    plt.xlabel(var)
    plt.ylabel('Proportion')
plt.tight_layout()
plt.show()
```

### Explore other Visualization Techniques

```python
# a. Display Distributions of Continuous Variables
```

```python
# Define continuous variables
continuous_vars = ['BMI', 'MentHlth', 'PhysHlth']

# Plotting histograms
plt.figure(figsize=(14, 6))

# Histograms
for i, var in enumerate(continuous_vars, 1):
    plt.subplot(1, 3, i)
    sns.histplot(df[var], kde=True, color='salmon', bins=20)
    plt.title(f'Distribution of {var}')
    plt.xlabel(var)
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
# b. Heatmaps and Correlation Plots: Explore Variable Relationships
```

```python
# Select relevant columns for correlation analysis
corr_columns = ['HighBP', 'HighChol','CholCheck','BMI', 'Smoker',
                'Stroke', 'Diabetes', 'PhysActivity', 'Fruits', 'Veggies',
                'HvyAlcoholConsump', 'AnyHealthcare', 'GenHlth',
                'MentHlth', 'PhysHlth', 'DiffWalk', 'Age', 'Income',
                'HeartDiseaseorAttack']

# Calculate correlation matrix
corr_matrix = df[corr_columns].corr()

# Plot heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix')
plt.show()
```

```python
# c. Scatter plots for relationships with HeartDiseaseorAttack
```

```python
plt.figure(figsize=(14, 10))
for i, var in enumerate(continuous_vars, 1):
    plt.subplot(2, 2, i)
    sns.scatterplot(x=var, y='HeartDiseaseorAttack', data=df, color='orange', alpha=0.6)
    plt.title(f'{var} vs Heart Disease')
    plt.xlabel(var)
    plt.ylabel('Heart Disease')
plt.tight_layout()
plt.show()
```

```python
# Line plots for trends over time or ordered categories
```

```python
# For trends by Age
age_groups = df['Age'].unique()
plt.figure(figsize=(12, 6))
for age_group in age_groups:
    subset = df[df['Age'] == age_group]
    mean_hd = subset.groupby('Age')['HeartDiseaseorAttack'].mean()
    plt.plot(mean_hd.index, mean_hd.values, marker='o', label=f'Age Group {age_group}')
plt.title('Average Heart Disease by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Average Heart Disease')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# d. Violin plots for distribution comparison
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='HeartDiseaseorAttack', y='BMI', data=df, palette='muted')
plt.title('BMI Distribution by Heart Disease Status')
plt.xlabel('Heart Disease or Attack')
plt.ylabel('BMI')
plt.show()
```

```python
# e. Pair plots for pairwise relationships
```

```python
pairplot_vars = ['BMI', 'MentHlth', 'PhysHlth', 'Fruits', 'Veggies', 'Age']
sns.pairplot(df[pairplot_vars], diag_kind='hist', corner=True)
plt.suptitle('Pairwise Relationships', y=1.02)
plt.show()
```
