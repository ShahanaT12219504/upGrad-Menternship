```python
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('heart_disease_health_indicators.csv')
```

### Univariate Analysis

```python
# a. Examine Distributions of Individual Variables
```

```python
# Plot histogram for Age
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], kde=True, bins=30, color='skyblue')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot histogram for BMI
plt.figure(figsize=(10, 6))
sns.histplot(df['BMI'], kde=True, bins=30, color='lightgreen')
plt.title('Distribution of BMI')
plt.xlabel('BMI')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot histogram for Mental Health Days (MentHlth)
plt.figure(figsize=(10, 6))
sns.histplot(df['MentHlth'], kde=True, bins=30, color='salmon')
plt.title('Distribution of Mental Health Days')
plt.xlabel('Mental Health Days')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot histogram for Physical Health Days (PhysHlth)
plt.figure(figsize=(10, 6))
sns.histplot(df['PhysHlth'], kde=True, bins=30, color='purple')
plt.title('Distribution of Physical Health Days')
plt.xlabel('Physical Health Days')
plt.ylabel('Frequency')
plt.show()
```

```python
# b. Investigate Prevalence of Health Conditions
```

```python
# Define categorical columns
categorical_cols = ['HeartDiseaseorAttack', 'HighBP', 'HighChol','Diabetes']

# Plot count plots for categorical variables
for col in categorical_cols:
    plt.figure(figsize=(8, 6))
    sns.countplot(x=col, data=df, palette='pastel')
    plt.title(f'Distribution of {col}')
    plt.xlabel(f'{col}')
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.show()
```

```python
# c. Analyze Distribution of Heart Disease (Target Variable)
```

```python
# Plotting count plot for 'HeartDisease' column
plt.figure(figsize=(10, 6))
sns.countplot(x='HeartDiseaseorAttack', data=df, palette='coolwarm')
plt.title('Distribution of HeartDiseaseorAttack Cases')
plt.xlabel('HeartDiseaseorAttack')
plt.ylabel('Count')
plt.show()
```

```python
# Plot pie chart for Heart Disease (HeartDisease)
plt.figure(figsize=(8, 8))
df['HeartDiseaseorAttack'].value_counts().plot.pie(autopct='%1.1f%%', colors=['skyblue', 'lightcoral'], labels=['No', 'Yes'])
plt.title('Distribution of HeartDiseaseorAttack Cases')
plt.ylabel('')
plt.show()
```

### Bivariate Analysis

```python
# a. Explore Relationships with Heart Disease

# Compare distribution of key variables between individuals with and without heart disease
```

```python
# Plotting for High Blood Pressure (HighBP)
plt.figure(figsize=(10, 6))
sns.countplot(x='HighBP', hue='HeartDiseaseorAttack', data=df, palette='pastel')
plt.title('Distribution of High Blood Pressure by HeartDiseaseorAttack')
plt.xlabel('High Blood Pressure')
plt.ylabel('Count')
plt.legend(title='HeartDiseaseorAttack', labels=['No', 'Yes'])
plt.show()
```

```python
# Plotting for High Cholesterol (HighChol)
plt.figure(figsize=(10, 6))
sns.countplot(x='HighChol', hue='HeartDiseaseorAttack', data=df, palette='muted')
plt.title('Distribution of High Cholesterol by HeartDiseaseorAttack')
plt.xlabel('High Cholesterol')
plt.ylabel('Count')
plt.legend(title='HeartDiseaseorAttack', labels=['No', 'Yes'])
plt.show()
```

```python
# Plotting for BMI (Body Mass Index)
plt.figure(figsize=(10, 6))
sns.boxplot(x='HeartDiseaseorAttack', y='BMI', data=df, palette='Set3')
plt.title('BMI Distribution by HeartDiseaseorAttack')
plt.xlabel('HeartDiseaseorAttack')
plt.ylabel('BMI')
plt.show()
```

```python
# b. Visualize Correlations Between Variables
```

```python
# Correlation heatmap for continuous variables
plt.figure(figsize=(10, 8))
sns.heatmap(df[['BMI', 'MentHlth', 'PhysHlth', 'Age']].corr(), annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Heatmap of Continuous Variables')
plt.show()
```

```python
# c. Compare Heart Disease Across Demographic Groups

# Analyze prevalence of heart disease across different demographic groups
```

```python
# Pivot table to compare heart disease by gender and education level (example)
heart_disease_pivot = pd.pivot_table(df, index='Sex', columns='Education', values='HeartDiseaseorAttack', aggfunc='mean')
plt.figure(figsize=(10, 6))
sns.heatmap(heart_disease_pivot, annot=True, cmap='YlGnBu', vmin=0, vmax=1)
plt.title('HeartDiseaseorAttack Prevalence by Gender and Education')
plt.xlabel('Education Level')
plt.ylabel('Sex')
plt.show()
```
